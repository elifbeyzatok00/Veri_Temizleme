```python
import pandas as pd
```

```python
dosya1_yolu = '/content/veri_setleri/form1.xlsx'  # İlk Excel dosyasının yolunu belirtin
dosya2_yolu = '/content/veri_setleri/form2.xlsx'  # İkinci Excel dosyasının yolunu belirtin
```

```python
# Excel dosyalarını DataFrame'e yükleme
veri_df1 = pd.read_excel(dosya1_yolu)
veri_df2 = pd.read_excel(dosya2_yolu)
```

```python
# DataFrameleri birleştirme
birlesik_df = pd.concat([veri_df1, veri_df2])
```

```python
birlesik_df
```

```python
# Tekrar eden satırları silme
temizlenmis_df = birlesik_df.drop_duplicates()
```

```python
temizlenmis_df
```

```python
# Temizlenmiş verileri yeni bir Excel dosyasına kaydetme
cikti_dosyasi = 'birlesik_veriler.xlsx'
temizlenmis_df.to_excel(cikti_dosyasi, index=False)
```

```python
print("Veriler birleştirildi ve temizlenmiş dosya kaydedildi: ", cikti_dosyasi)
```
